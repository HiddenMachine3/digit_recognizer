```python
import tensorflow as tf
from tensorflow.keras.utils import to_categorical
import keras
```

```python
model = tf.keras.models.Sequential(
    [
    tf.keras.layers.Conv2D(filters=6, kernel_size=(5, 5),padding="same", activation='sigmoid', input_shape=(28, 28, 1)),
    tf.keras.layers.AvgPool2D(pool_size=(2,2),strides=2,data_format='channels_last'),
    tf.keras.layers.Conv2D(filters=16, kernel_size=(5, 5),padding="valid", activation='sigmoid'),
    tf.keras.layers.AvgPool2D(pool_size=(2,2),strides=2,data_format='channels_last'),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(120, activation='sigmoid'),
    tf.keras.layers.Dense(84, activation='sigmoid'),
    tf.keras.layers.Dense(10, activation='softmax')
    ]
)
```

```python
model.compile(
    loss=keras.metrics.categorical_crossentropy,
    optimizer=keras.optimizers.Adam(),
    metrics=["accuracy"],
)
```

```python
handwritten_digits =  tf.keras.datasets.mnist.load_data()
(train_imgs, train_lbls), (test_imgs, test_lbls) = handwritten_digits
train_lbls = to_categorical(train_lbls)
test_lbls = to_categorical(test_lbls)
```

```python
def display_ascii_with_color(array):
    ascii_array = []
    for row in array:
        ascii_row = []
        for element in row:
            brightness = int(element / 16) + 232
            ascii_row.append(f'\033[48;5;{brightness}m \033[0m')
        ascii_array.append(''.join(ascii_row))
    return '\n'.join(ascii_array)

ascii_output = display_ascii_with_color(train_imgs[0])
print(train_lbls[0].argmax())
print(ascii_output)
```

```python
model.fit(train_imgs, train_lbls, epochs=10, validation_data=(test_imgs, test_lbls))
```

```python
import numpy as np
ascii_output = display_ascii_with_color(test_imgs[0])
actual_number = test_lbls[0].argmax()
predicted_number = model.predict(np.array([test_imgs[0]])).argmax()
print("actual :", actual_number,", predicted :", predicted_number)
print(ascii_output)
```
