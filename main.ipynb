```python
import tensorflow as tf
from tensorflow.keras.utils import to_categorical
from pprint import pprint
import numpy as np
```

```python
handwritten_digits =  tf.keras.datasets.mnist.load_data()
(train_imgs, train_lbls), (test_imgs, test_lbls) = handwritten_digits
```

```python
train_imgs[0]
```

```python
def display_ascii_with_color(array):
    ascii_array = []
    for row in array:
        ascii_row = []
        for element in row:
            brightness = int(element / 16) + 232  # Calculate brightness value for ANSI color
            ascii_row.append(f'\033[48;5;{brightness}m \033[0m')  # ANSI color output with brightness
        ascii_array.append(''.join(ascii_row))
    return '\n'.join(ascii_array)
```

```python
ascii_output = display_ascii_with_color(train_imgs[0])
print(train_lbls[0])
print(ascii_output)
```

```python
train_lbls = to_categorical(train_lbls)
test_lbls = to_categorical(test_lbls)
```

```python
model = tf.keras.models.Sequential(
    [
        tf.keras.layers.Flatten(),
        tf.keras.layers.Dense(30, activation="sigmoid"),
        tf.keras.layers.Dense(10, activation="softmax"),
    ]
)

model.compile(
    optimizer="adam",
    loss=tf.keras.losses.CosineSimilarity(),
    metrics=["accuracy"],
)

model.fit(x=train_imgs, y=train_lbls, epochs=30)
```

```python
ascii_output = display_ascii_with_color(test_imgs[0])
actual_number = test_lbls[0].argmax()
predicted_number = model.predict(np.array([test_imgs[0]])).argmax()
print("actual :", np.argmax(test_lbls[0]),", predicted :", predicted_number)
print(ascii_output)
```
